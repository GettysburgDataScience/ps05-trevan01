````python
# Female Survivors DF
columns_to_keep = ['Age', 'Name', 'Sex']
fem_survived_df = titanic_df.query('Survived == 1 and Sex == 0')[columns_to_keep]
fem_survived_df.head()
````

````python
# Male Survivors DF
columns_to_keep = ['Age', 'Name', 'Sex']
male_survived_df = titanic_df.query('Survived == 1 and Sex == 1')[columns_to_keep]
male_survived_df.head()
````

````python
# Survivor DF for both sexes
survivor_df = pd.merge(fem_survived_df,male_survived_df,
                       how = 'outer')
survivor_df.tail()
````

````python
# Female Non-Survivors DF
columns_to_keep = ['Age', 'Name', 'Sex']
fem_nonsurvived_df = titanic_df.query('Survived == 0 and Sex == 0')[columns_to_keep]
fem_nonsurvived_df.head()
````

````python
# Male Non-Survivors DF
columns_to_keep = ['Age', 'Name', 'Sex']
male_nonsurvived_df = titanic_df.query('Survived == 0 and Sex == 1')[columns_to_keep]
male_nonsurvived_df.head()
````

````python
# Nonsurvivor DF for both sexes
nonsurvivor_df = pd.merge(fem_nonsurvived_df, male_nonsurvived_df,
                       how = 'outer')
nonsurvivor_df.head()
````

````python
plt.figure(figsize = (12, 9))

plt.hist(male_nonsurvived_df['Age'], bins=20, alpha = 1, label='Male Nonsurvivors', color='red', histtype='stepfilled')
plt.hist(fem_survived_df['Age'], bins=20, alpha = 0.7, label='Female Survivors', color='green', histtype='stepfilled')
plt.hist(male_survived_df['Age'], bins=20, alpha = 0.7, label='Male Survivors', color='blue', histtype='stepfilled')
plt.hist(fem_nonsurvived_df['Age'], bins=20, alpha = 0.7, label='Female Nonsurvivors', color='purple', histtype='stepfilled')

plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Age Distribution by Sex and Survivorship', fontsize=16, fontweight='bold')
plt.legend()
````
